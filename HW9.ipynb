```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
with sqlite3.connect('populationn.db') as con:
    df = pd.read_sql('SELECT * FROM population', con=con)

# Clean column names
df = df.drop(columns='id')
df.columns = ['First Name', 'Last Name', 'Email', 'Gender', 'Salary', 'State']
```
